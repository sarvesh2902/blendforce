```python
import pandas as pd
```

```python
employees = pd.read_csv("employee_data.csv")
```

```python
employees.head()
```

```python
selected_columns = employees[['Name', 'Surname', 'Division', 'Gender', 'Ethnicity', 'Sexual_Orientation', 'LGBTQ', 'Indigenous', 'Disability', 'Age' ,'Mobile Number', 'Email']]
```

```python
selected_columns.head()
```

```python
selected_columns['Full Name'] = selected_columns['Name'] + ' ' + selected_columns['Surname']
```

```python
selected_columns
```

```python
selected_columns = selected_columns.drop(['Name', 'Surname'], axis=1)
```

```python
selected_columns
```

```python
selected_columns = selected_columns.rename(columns={'Full Name': 'name', 'Email': 'email','Mobile Number': 'mobile', 'Gender': 'gender','Ethnicity': 'ethnicity', 'Sexual_Orientation': 'sexualOrientation','LGBTQ': 'lgbtq', 'Indigenous': 'indigenous'})
```

```python
selected_columns
```

```python
selected_columns = selected_columns.rename(columns={'Division': 'department', 'Disability': 'disability','Age': 'age'})
```

```python
selected_columns
```

```python
import numpy as np

positions = {
    'IT': ['software developer', 'IT consultant', 'network engineer', 'systems analyst', 'database administrator', 'cybersecurity specialist', 'cloud architect'],
    'Marketing': ['marketing manager', 'digital marketer', 'social media specialist', 'brand strategist', 'market researcher'],
    'Finance': ['financial analyst', 'accountant', 'financial controller', 'investment analyst', 'tax specialist'],
    'HR': ['HR manager', 'recruitment specialist', 'compensation and benefits analyst', 'training and development coordinator', 'employee relations consultant'],
    'RD': ['research scientist', 'R&D engineer', 'innovation manager', 'product development specialist', 'technology analyst'],
    'Sales': ['sales manager', 'account executive', 'business development representative', 'sales operations analyst', 'customer success manager'],
    'Operations': ['operations manager', 'supply chain analyst', 'logistics coordinator', 'process improvement specialist', 'quality assurance engineer']
}
selected_columns['position'] = selected_columns['department'].apply(lambda x: np.random.choice(positions[x]))
```

```python
selected_columns
```

```python
selected_columns.to_csv('employee_final.csv', index=False)
```
